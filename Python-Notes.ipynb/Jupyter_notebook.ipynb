Lambda expression

```python
def time2(var):
    return var * 3
```

```python
time2(4)
```

```python
lambda var: var * 2
```

map and Filter

```python
sqr = [1, 2 , 3, 9, 8 , 12]
```

```python
map(time2, sqr)
```

```python
list(map(time2,sqr))
```

```python
list(map(lambda var: var*3, sqr))
```

filter

```python
filter(lambda item: item%2 == 0, sqr)
```

```python
list(filter(lambda item: item%2 == 0,sqr))
```
